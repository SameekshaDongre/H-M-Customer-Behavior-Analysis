```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
articles = pd.read_csv("articles.csv")
customers = pd.read_csv("customers.csv")
transactions = pd.read_csv("transactions_train.csv")
```

```python
def display_data_info():
    print("Dataset Information")
    datasets = {"Articles": articles, "Customers": customers, "Transactions": transactions}

    for name, df in datasets.items():
        print(f"\n{name} Data Summary:")
        print(df.info())
        print(f"Missing Values:\n{df.isnull().sum()}\n")

display_data_info()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=transactions['sales_channel_id'])
plt.title("Online vs Store Sales")
plt.xlabel("Sales Channel (1 = Store, 2 = Online)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=customers['Active'].fillna(0))
plt.title("Active Customer Distribution")
plt.xlabel("Active (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=customers['FN'].fillna(0))
plt.xticks([0, 1], ["Not Subscribed (0)", "Subscribed (1)"])
plt.title("Fashion News (FN) Subscription")
plt.xlabel("Subscription Status")
plt.ylabel("Count")
plt.show()
```
